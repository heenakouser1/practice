Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.


```python
#Glyph is used to shade a region with particular color, 
# we have circle, path glyphs


#example
# The patch glyph shades a region of space in a particular color. We use the patch() 
# method to develop a single patch and patches() method to develop multiple patches.
from bokeh.plotting import figure, show, output_file
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 5, 2]
p = figure(title='Patch Glyph',
           plot_height=400, plot_width=600,
           x_axis_label='x', y_axis_label='y',
           toolbar_location=None)
p.patch(x, y, fill_color="blue", line_color='black',
        alpha=0.5, line_width=2)
show(p)
output_file("ex.html")
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?


```python
from bokeh.models import axes
x=[1,2,3,4,5]
y=[4,5,7,8,9]
output_file("line.html")
p=figure(title='line graph')
p.xaxis.axis_label='x-axis'
p.yaxis.axis_label='y-axis'
p.scatter(x,y,size=20,fill_color='red',legend_label='its data point')
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?


A Bokeh server is a Python application that allows you to create interactive web-based visualizations that can be updated in real time. By using the Bokeh server, you can create a plot or dashboard that responds to user input or data changes and updates the visualization automatically, without the need to refresh the page.
To create a Bokeh server application, you need to define a function that returns a Layout object (or a subclass of Layout, such as column or row). The Layout object represents the structure of the web page, and can contain multiple plots and widgets.

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?


If an application is running on a Bokeh server that makes it available at some URL,
you will typically want to embed the entire application in a web page. This way, the page will create a new session and display it to the user every time it loads.

You can achieve this with the server_document() function. This function accepts the URL to a Bokeh server application and returns a script that embeds a new session from that server every time the script executes.

Here is an example of the server_document() function in use:

```python
from bokeh.embed import server_document
script = server_document("https://demo.bokeh.org/sliders")
```
